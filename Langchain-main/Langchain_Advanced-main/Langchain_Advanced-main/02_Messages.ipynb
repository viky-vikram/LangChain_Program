# 2. Messages

**Messages** are how we talk to a chat model in LangChain.

Every chat is a **list of messages**, and each message has a **role** (who is speaking).

## The main message types

| Message | Who | Used for |
|---------|-----|----------|
| **SystemMessage** | You (setup) | Instructions for how the AI should behave |
| **HumanMessage** | The user | What the person says |
| **AIMessage** | The AI | The model's reply |

## Simple real-life example

Think of a play script:
- **SystemMessage** = the director's notes ("act friendly and short").
- **HumanMessage** = the user's line.
- **AIMessage** = the AI's line back.

## Step 1: Import the message types

```python
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()
llm = ChatOpenAI(model="gpt-4o-mini")
print("Ready!")
```

## Step 2: Send a list of messages

The **SystemMessage** sets the behavior. The **HumanMessage** is the question.

We pass them as a list to `invoke()`.

```python
messages = [
    SystemMessage(content="You are a friendly teacher. Keep answers very short."),
    HumanMessage(content="What is a variable in Python?")
]

response = llm.invoke(messages)
print(response.content)
```

## Step 3: The reply is an AIMessage

The model's answer comes back as an **AIMessage**. The text is in `.content`.

```python
print("Type of reply:", type(response).__name__)
print("Reply text:", response.content)
```

## Step 4: Holding a conversation (memory)

The model does not remember on its own. To continue a chat, we **add each message to the list** and send the whole list again.

```python
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="My name is Anjali.")
]

# First reply
reply = llm.invoke(messages)"What is my name?
print("AI:", reply.content)

# Add the AI's reply, then ask a follow-up
messages.append(reply)
messages.append(HumanMessage(content="))

reply = llm.invoke(messages)
print("AI:", reply.content)
```

## Key points to remember

- A chat is a **list of messages**.
- **SystemMessage** = instructions for the AI.
- **HumanMessage** = what the user says.
- **AIMessage** = the model's reply (text in `.content`).
- To keep a conversation, add each message to the list and resend it.
- The model has no memory on its own; the message list IS the memory.
