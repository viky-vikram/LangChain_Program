# 1. Model Integration

**Model integration** means **connecting LangChain to an AI model** so we can send it messages and get replies.

The main model in this course is **OpenAI's GPT**, but LangChain can connect to many providers. Later in this notebook we also show **Ollama**, **Claude**, and **Google Gemini**.

## Step 1: Install and set your key

In the terminal:
```
pip install langchain-openai
```

Then set your API key (Windows PowerShell):
```
$env:OPENAI_API_KEY = "your-key-here"
```

> The key is how OpenAI knows it's you. Never share it or put it in your code.

## Step 2: Create the model

We use `ChatOpenAI` and pick a model. `gpt-4o-mini` is a fast, low-cost model that's great for learning.

````python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(model="gpt-4o-mini",)

print("Model is ready!")
````

## Step 3: Send a message with `invoke()`

`invoke()` sends text to the model and returns the reply.

The reply is a message object; the text is in `.content`.

````python
response = llm.invoke("What is Python in one sentence?")

print(response.content)
````

## Common settings

You can control how the model behaves:

- **`model`** -> which GPT model to use (e.g. `gpt-4o-mini`, `gpt-4o`).
- **`temperature`** -> how creative the reply is (0 = focused, 1 = creative).

Example below makes replies more focused.

````python
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0   # 0 = focused and predictable
)

response = llm.invoke("Give me 3 fun facts about space.")
print(response.content)
````

## Streaming the reply (optional)

Instead of waiting for the full answer, we can print it **word by word** as it arrives using `stream()`.

````python
for chunk in llm.stream("Count from 1 to 5 slowly."):
    print(chunk.content, end="", flush=True)
````

# Connecting to OTHER AI providers

The best part of LangChain: **only the first line (creating the model) changes**.

Once you have an `llm`, everything else is the same -- `invoke()`, `stream()`, messages, tools all work identically. So you can swap providers easily.

## Option A: Ollama (run models on YOUR computer)

**Ollama** lets you run AI models **locally** -- free, private, and no API key needed.

### Setup
1. Download and install Ollama from **https://ollama.com**
2. In the terminal, download a model:
   ```
   ollama pull llama3
   ```
3. Install the LangChain connector:
   ```
   pip install langchain-ollama
   ```

````python
from langchain_ollama import ChatOllama

# Runs on your own computer -- no API key needed
llm = ChatOllama(model="llama3")

response = llm.invoke("What is Python in one sentence?")
print(response.content)
````

## Option B: Claude (Anthropic API)

**Claude** is made by Anthropic. You need an Anthropic API key.

### Setup
```
pip install langchain-anthropic
```

Set your key (Windows PowerShell):
```
$env:ANTHROPIC_API_KEY = "your-key-here"
```

````python
from langchain_anthropic import ChatAnthropic

llm = ChatAnthropic(model="claude-opus-4-8")

response = llm.invoke("What is Python in one sentence?")
print(response.content)
````

## Option C: Google Gemini (Google API)

**Gemini** is made by Google. You need a Google API key (get one from Google AI Studio).

### Setup
```
pip install langchain-google-genai
```

Set your key (Windows PowerShell):
```
$env:GOOGLE_API_KEY = "your-key-here"
```

````python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")

response = llm.invoke("What is Python in one sentence?")
print(response.content)
````

## Quick comparison

| Provider | Install | Create the model | API key needed? |
|----------|---------|------------------|-----------------|
| **OpenAI** | `langchain-openai` | `ChatOpenAI(model="gpt-4o-mini")` | Yes (`OPENAI_API_KEY`) |
| **Ollama** | `langchain-ollama` | `ChatOllama(model="llama3")` | No (runs locally) |
| **Claude** | `langchain-anthropic` | `ChatAnthropic(model="claude-opus-4-8")` | Yes (`ANTHROPIC_API_KEY`) |
| **Google** | `langchain-google-genai` | `ChatGoogleGenerativeAI(model="gemini-2.0-flash")` | Yes (`GOOGLE_API_KEY`) |

After this line, `invoke()`, `stream()`, messages, and tools work **exactly the same** for all of them.

## Key points to remember

- Model integration = connecting LangChain to an AI model.
- OpenAI: `ChatOpenAI(model="gpt-4o-mini")` + `OPENAI_API_KEY`.
- Ollama: `ChatOllama(model="llama3")` -> free, runs locally, no key.
- Claude: `ChatAnthropic(model="claude-opus-4-8")` + `ANTHROPIC_API_KEY`.
- Google: `ChatGoogleGenerativeAI(model="gemini-2.0-flash")` + `GOOGLE_API_KEY`.
- Only the model-creation line changes -- the rest of LangChain stays the same.
- `invoke("...")` returns a reply; read the text with `.content`.
