# 5. Retriever

A **Retriever** is the tool that **fetches the most relevant text** from the vector store when you ask a question.

It is the **final step** that connects your data to an AI model.

## Retriever vs Vector Store

- A **Vector Store** is the storage (the library shelf).
- A **Retriever** is the helper that goes and **brings back** the right books for your question.

A retriever is created *from* a vector store, and it has one simple job: **given a question, return the best matching chunks.**

## Step 1: Build a vector store (same as before)

We need the free embeddings + FAISS:
```
pip install langchain-huggingface sentence-transformers faiss-cpu
```

````python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

texts = [
    "Python is a programming language used for AI.",
    "Streamlit is used to build web apps quickly.",
    "LangChain helps connect AI models with your data.",
    "Dogs are friendly and loyal animals."
]

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_store = FAISS.from_texts(texts, embeddings)

print("Vector store ready!")
````

## Step 2: Create a Retriever

We turn the vector store into a retriever using `.as_retriever()`.

````python
retriever = vector_store.as_retriever(search_kwargs={"k": 2})  # return top 2 results

print("Retriever is ready!")
````

## Step 3: Ask a question

The retriever finds the most relevant chunks for our question.

````python
question = "What is used to build web apps?"

results = retriever.invoke(question)

print("Question:", question)
print("\nRelevant information found:")
for r in results:
    print("-", r.page_content)
````

## How this fits into a full AI app

Once the retriever finds the relevant text, we give that text **plus** the question to an AI model. The model then answers using *your* data.

Simple idea:
```
User question ➜ Retriever finds relevant text ➜ AI model reads it ➜ AI gives the answer
```

This is the heart of a **RAG (Retrieval-Augmented Generation)** app.

## Key points to remember

- A Retriever fetches the most relevant text for a question.
- It is created from a vector store with `.as_retriever()`.
- `k` controls how many results to return.
- Use `retriever.invoke(question)` to get the matching chunks.
- Retriever is the **bridge** between your data and the AI model in a RAG app.
