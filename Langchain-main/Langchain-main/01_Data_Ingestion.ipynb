# 1. Data Ingestion (Document Loaders)

**Data Ingestion** means **bringing data into our program** so the computer can use it.

In LangChain, we load data (like text files, PDFs, or websites) using **Document Loaders**.

This is the **first step** in building a RAG (Retrieval-Augmented Generation) app.

## Why do we need it?

An AI model does not know about *your* private data (your notes, your PDFs, your company files).

So first, we must **load that data** into the program. Then later we can search it and ask questions about it.

## Step 1: Install LangChain

Type this in the **terminal**:
```
pip install langchain langchain-community
```

## Step 2: Create a sample text file

The code below creates a simple text file that we will load.

````python
# Create a sample file to load
text = """LangChain is a framework for building AI apps.
It helps connect language models with your own data.
RAG means Retrieval-Augmented Generation."""

with open("sample.txt", "w") as f:
    f.write(text)

print("sample.txt file created!")
````

## Step 3: Load a text file with `TextLoader`

A **loader** reads the file and turns it into **Documents** that LangChain understands.

````python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("sample.txt")
documents = loader.load()

print("Number of documents:", len(documents))
print("Content:")
print(documents[0].page_content)
````

## Other common loaders

LangChain can load many types of data:

| Loader | Loads from |
|--------|-----------|
| `TextLoader` | A `.txt` file |
| `PyPDFLoader` | A PDF file |
| `CSVLoader` | A CSV (spreadsheet) file |
| `WebBaseLoader` | A website URL |

### Example: loading a PDF (needs `pip install pypdf`)
```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("mybook.pdf")
pages = loader.load()
print(pages[0].page_content)
```

## Key points to remember

- Data Ingestion = loading your data into the program.
- It is the **first step** of a RAG app.
- Use **Document Loaders** like `TextLoader`, `PyPDFLoader`, `CSVLoader`.
- `.load()` returns a list of **Documents**.
- Each document has `.page_content` (the actual text).
