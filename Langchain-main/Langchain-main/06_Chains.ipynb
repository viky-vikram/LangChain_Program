# 6. Chains

A **Chain** in LangChain joins steps together so they run **one after another**, like a pipeline.

The output of one step automatically becomes the input of the next step.

This is what lets you build real AI apps instead of writing lots of glue code by hand.

## Real-life analogy

Think of making a cup of tea — a chain of steps:

```
Boil water  ➜  Add tea leaves  ➜  Add milk & sugar  ➜  Pour into cup
```

Each step passes its result to the next one. A LangChain **chain** works the same way for AI tasks.

## The most common chain: Prompt ➜ Model ➜ Output

Three small steps joined together:

| Step | What it does |
|------|--------------|
| **Prompt** | Builds the question/instruction from your input |
| **Model** | The AI (OpenAI) that thinks and replies |
| **Output parser** | Cleans the reply into plain text |

In LangChain we join them with the **pipe** symbol `|`:
```
chain = prompt | model | output_parser
```

```
pip install langchain langchain-openai
```

## Step 2: Create the three pieces

````python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. Prompt: a template with a blank {topic} to fill in
prompt = ChatPromptTemplate.from_template("Explain {topic} in one simple sentence.")

# 2. Model: the OpenAI chat model
model = ChatOpenAI(model="gpt-4o-mini")

# 3. Output parser: turn the model's reply into plain text
output_parser = StrOutputParser()

print("All three pieces are ready!")
````

## Step 3: Join them into a chain

We connect the pieces with `|`. Read it left to right, like an arrow.

````python
chain = prompt | model | output_parser

print("Chain is ready!")
````

## Step 4: Run the chain

We call `.invoke()` and pass values for the blanks in the prompt. The whole pipeline runs automatically.

````python
answer = chain.invoke({"topic": "Python"})

print(answer)
````

## Bonus: a slightly longer chain

You can keep adding steps. Here we make a chain that first writes a tip, and the `|` still joins everything in order.

````python
tip_prompt = ChatPromptTemplate.from_template(
    "Give one short, friendly tip for a beginner learning {skill}."
)

tip_chain = tip_prompt | model | output_parser

print(tip_chain.invoke({"skill": "Python"}))
````

## Key points to remember

- A **Chain** joins steps so they run one after another, like a pipeline.
- The output of one step becomes the input of the next.
- The most common chain is **Prompt ➜ Model ➜ Output parser**.
- We join steps with the **pipe** symbol `|`.
- We run the chain with `.invoke({...})`, passing values for the prompt's blanks.
- Build the chain **once**, reuse it many times with different inputs.
- Chains are the backbone of real LangChain apps (including RAG).
