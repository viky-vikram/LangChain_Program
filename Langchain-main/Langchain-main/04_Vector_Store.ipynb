# 4. Vector Store

A **Vector Store** is a special **database** that saves embeddings (the number vectors).

It lets us **search by meaning** very quickly.

This is the **fourth step** in a RAG app.

## Why do we need it?

After we turn text into vectors, we need a place to **store** them and **search** them.

A normal database searches for exact words. A vector store searches for **meaning**.

### Simple example
If you search "car", a vector store can also find "vehicle" or "automobile" — because they mean the same thing.

## Step 1: Install a free vector store (FAISS)

FAISS is a free, fast vector store made by Facebook.

In the terminal:
```
pip install faiss-cpu
```
(You also need an embedding model. Free: `pip install langchain-huggingface sentence-transformers`. OpenAI: `pip install langchain-openai`.)

## Step 2: Prepare some text and the embedding model

A vector store always needs an **embedding model** to turn text into vectors. You can use the **free** one or the **OpenAI** one — the vector store works the same either way.

````python
from langchain_huggingface import HuggingFaceEmbeddings

# Our small knowledge base (a list of sentences)
texts = [
    "Python is a programming language.",
    "Streamlit is used to build web apps.",
    "Dogs are friendly animals.",
    "Cats like to sleep a lot."
]

# Free, local embedding model
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
print("Ready!")
````

### Using OpenAI embeddings instead (optional)

Want higher-quality (paid) embeddings? Just swap the model. Set `OPENAI_API_KEY` first, then:
```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```
Everything below (storing and searching) stays **exactly the same** — FAISS does not care which embedding model made the vectors.

## Step 3: Store the text in FAISS

`FAISS.from_texts()` automatically converts each sentence into a vector and saves it.

````python
from langchain_community.vectorstores import FAISS

vector_store = FAISS.from_texts(texts, embeddings)

print("All texts are stored in the vector store!")
````

## Step 4: Search by meaning

We ask a question, and the store returns the **most similar** sentences.

````python
query = "Tell me about pets"

results = vector_store.similarity_search(query, k=2)   # k=2 means top 2 matches

print("Top matches for:", query)
for r in results:
    print("-", r.page_content)
````

````python
vector_store.save_local("my_store")
````

````python
vector_store = FAISS.load_local("my_store", embeddings, allow_dangerous_deserialization=True)
````

## Step 5: Save and load the store

You can save the store to disk and load it later, so you don't rebuild it every time.
```python
vector_store.save_local("my_store")

# load it back later
vector_store = FAISS.load_local("my_store", embeddings, allow_dangerous_deserialization=True)
```

## Key points to remember

- A Vector Store is a database for embeddings (vectors).
- It searches by **meaning**, not exact words.
- It is the **fourth step** in a RAG app.
- **FAISS** is a free, popular vector store.
- `FAISS.from_texts()` stores the data; `similarity_search()` finds matches.
- It works with **any** embedding model — free HuggingFace or paid **OpenAI** (`text-embedding-3-small`). Only the embedding line changes.
- Other vector stores: **Chroma**, **Pinecone**.
