# 3. Embeddings

**Embeddings** turn text into **numbers** (a list of numbers called a *vector*).

Computers cannot understand words, but they understand numbers. Embeddings let the computer compare meaning using math.

This is the **third step** in a RAG app.

## The simple idea

Words with **similar meaning** get **similar numbers**.

- "king" and "queen" → numbers close to each other.
- "king" and "banana" → numbers far apart.

This is how the computer knows two pieces of text "mean" almost the same thing.

## Two ways to make embeddings

| Way | Model | Cost | API key? |
|-----|-------|------|----------|
| **Free / local** | HuggingFace `all-MiniLM-L6-v2` | Free | No |
| **OpenAI (paid)** | `text-embedding-3-small` | Paid | Yes |

We will show **both**. They work almost the same way — only the model line changes.

## Option A: Free local model (HuggingFace)

We use a **free, local** model from HuggingFace (no paid API key needed).

In the terminal:
```
pip install langchain-huggingface sentence-transformers
```

> The first time you run it, it downloads a small model. This may take a minute.

````python
from langchain_huggingface import HuggingFaceEmbeddings

# A small, free, popular embedding model
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

print("Embedding model is ready!")
````

## Option B: OpenAI embeddings (paid)

OpenAI gives very high-quality embeddings. They cost money and need an API key.

In the terminal:
```
pip install langchain-openai
```

Then set your key (Windows PowerShell):
```
$env:OPENAI_API_KEY = "your-key-here"
```

````python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")


# A small, cheap, high-quality OpenAI embedding model
openai_embeddings = OpenAIEmbeddings(model="text-embedding-3-small", api_key = api_key)

print("OpenAI embedding model is ready!")
````

### Notice: the code is almost the same!

| | Free local | OpenAI |
|---|---|---|
| Import | `from langchain_huggingface import HuggingFaceEmbeddings` | `from langchain_openai import OpenAIEmbeddings` |
| Create | `HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")` | `OpenAIEmbeddings(model="text-embedding-3-small")` |
| Use | `.embed_query(text)` | `.embed_query(text)` |

Only the **import** and the **create** line change. Everything after that is the same.

> In the examples below we use the **free** `embeddings` so it runs without a key. To use OpenAI instead, just swap `embeddings` for `openai_embeddings`.

## Turn text into numbers

````python
text = "I love programming in Python."

vector = embeddings.embed_query(text)

print("How many numbers in this vector?", len(vector))
print("First 5 numbers:", vector[:5])
````

## Compare meanings

Let's see that similar sentences get similar vectors.

We use a simple math measure (**cosine similarity**): closer to **1** = more similar.

````python
import numpy as np

def similarity(a, b):
    a, b = np.array(a), np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

v1 = embeddings.embed_query("I like dogs")
v2 = embeddings.embed_query("I love puppies")
v3 = embeddings.embed_query("The car is fast")

print("dogs vs puppies (similar):", round(similarity(v1, v2), 2))
print("dogs vs car   (different):", round(similarity(v1, v3), 2))
````

## Key points to remember

- Embeddings convert text into a list of numbers (a **vector**).
- Similar meanings → similar numbers.
- It is the **third step** in a RAG app.
- `embed_query()` turns one sentence into a vector.
- We compare vectors with **cosine similarity** (closer to 1 = more similar).
- **Free option:** HuggingFace `all-MiniLM-L6-v2` (no API key).
- **Paid option:** OpenAI `text-embedding-3-small` (needs `OPENAI_API_KEY`).
- Only the model line changes — the rest of the code stays the same.
