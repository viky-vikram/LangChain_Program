# 2. Text Splitter

A **Text Splitter** breaks a large piece of text into **smaller pieces** called **chunks**.

This is the **second step** in a RAG app.

## Why split the text?

- AI models can only read a **limited amount of text** at a time.
- Smaller chunks give **more accurate** search results.
- It is easier to find the *exact* part that answers a question.

### Simple example
Imagine a 100-page book. Instead of searching the whole book, we cut it into small paragraphs. Then we only read the paragraph that matches the question.

## Two important settings

- **`chunk_size`** → how big each chunk is (number of characters).
- **`chunk_overlap`** → how much text repeats between chunks (so we don't lose meaning at the edges).

A B C D E F G

[A B C]  [B C D] [C D E] [D E F] [E F G]

## Step 1: Our long text

````python
long_text = """Python is a programming language. It is easy to learn.
Streamlit is used to build web apps. LangChain helps build AI apps.
RAG combines search with AI models. Embeddings turn text into numbers.
Vector stores save those numbers. Retrievers find the best matches."""

print("Total characters:", len(long_text))
````

## Step 2: Split the text

We use `RecursiveCharacterTextSplitter` — the most common and beginner-friendly splitter.

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=80,      # each chunk is about 80 characters
    chunk_overlap=10    # 10 characters repeat between chunks
)

chunks = splitter.split_text(long_text)

print("Number of chunks:", len(chunks))
for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}:")
    print(chunk)
````

## Splitting Documents (from a loader)

If your text came from a Document Loader, use `split_documents()` instead.
```python
from langchain_community.document_loaders import TextLoader

documents = TextLoader("sample.txt").load()
chunks = splitter.split_documents(documents)
```

## Key points to remember

- A Text Splitter cuts large text into small **chunks**.
- It is the **second step** in a RAG app.
- `chunk_size` = size of each chunk; `chunk_overlap` = repeated text between chunks.
- `RecursiveCharacterTextSplitter` is the most common splitter.
- Use `split_text()` for plain text, `split_documents()` for loaded documents.
